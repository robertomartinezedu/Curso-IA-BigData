#1 - EXPLICACIÓN Y CONTEXTUALIZACIÓN DE LOS DATOS
Campos del dataset _diamonds_ y su significado:


*   ```carat``` : Peso del diamante en gramos.
*   ```cut``` : Calidad del corte, distinguiendo entre cinco categorías:
    * Fair
    * Good
    * Very Good
    * Premium
    * Ideal
*   ```color``` : Calificación del color considerando el nivel de transparencia, con valores de mayor a menor calidad: D, E, F, G, H, I, J.
*   ```clarity``` : Representa la claridad del diamante, que es mayor cuantas menos imperfecciones tiene. Este campo abarca desde la mejor claridad, representada con el valor IF(Internally Flawless - sin imperfecciones internas, hasta el peor nivel de claridad, que es representado por el valor I3(imperfecciones visibles a simple vista). Los valores posibles de mejor a peor calidad son los siguientes: FL, IF, VVS1, VVS2, VS1, VS2, SI1, SI2, I1, I2, I3.
*   ```table``` : Indica la calidad del corte, representando una correlación entre el ancho de la parte superior del diamante y su anchura máxima, expresado en un porcentaje. Este valor es relevante en cuanto a que afecta a cómo se va a comportar la luz una vez incide a través de la parte superior de la gema y se refleja internamente.

   ![Tabla](https://www.brilliance.com/front/img/table_illustration.1d5c5a3f.jpeg)
*   ```price``` : Precio en dólares estadounidenses.
*   ```x``` : Longitud en milímetros.
*   ```y``` : Anchura en milímetros.
*   ```z``` : Profundidad en milímetros.
*   ```depth``` : Expresa una relación porcentual entre el ancho del diamante y la altura, medida desde la mesa hasta el vértice inferior.

   ![Depth](https://www.brilliance.com/front/img/depth_illustration.3cfc28e7.jpeg)


 Tras la primera inspección del dataframe se puede observar que todos sus campos contienen datos y no va a haber presencia de nulos, lo que favorece en gran medida el análisis de los mismos.

 Para enriquecer el análisis inicial se realiza un primer vistazo a nivel estadístico del dataframe utilizando la función _describe_ que ofrece la librería _seaborn_, así como calculando el coeficiente de de variación de cada columna.









````python
df.describe()
````

````python
df_estadisticas = df.describe()

media = df_estadisticas.loc['mean']
desviacion_estandar = df_estadisticas.loc['std']

porcentaje_desviacion = ((desviacion_estandar / media) * 100).round(2)
porcentaje_desviacion = porcentaje_desviacion.astype(str) + ' %'
print(porcentaje_desviacion)
````

#3 - ANÁLISIS DE LAS FILAS
Con el siguiente código se muestran todos los datos que tengan un cero en alguna de las tres columnas y el total de los mismos.



````python
df.loc[(df['x']==0) | (df['y']==0) | (df['z']==0)]
````

````python
print("Total de diamantes con datos inválidos: " + str(len(df[(df['x']==0) | (df['y']==0) | (df['z']==0)])))
````

Se puede observar que hay 20 diamantes con datos erróneos, así que se procede a eliminarlos.

````python
df = df[(df[['x','y','z']] != 0).all(axis=1)]
````

Con respecto a las columnas con datos categóricos, que son ```cut```, ```color``` y ```clarity``` se ha decidido respetar sus datos para realizar un análisis correlativo con el resto de propiedades del diamante y observar si pueden ser relevantes y ofrecer algún tipo de información.
Como se observaba en la primera inspección de los datos, estas tres columnas corresponden a categorías, y podemos visualizar todos los datos diferentes que contienen de la siguiente forma:

````python
print(df['cut'].unique())
print(df['color'].unique())
print(df['clarity'].unique())
````

Hasta el momento se han analizado aspectos al respecto de las diferentes columnas existentes, pero también merece la pena prestarle atención a las filas, pues al tratarse de un gran volumen de datos, puede haber duplicados.
Se puede comprobar con la función `duplicated()` de la librería _pandas_.

````python
duplicados = df.duplicated()
num_duplicados = duplicados.sum()
print(f"Número de filas duplicadas: {num_duplicados}")
````

Tras ejecutar el código se observa que hay presentes 803 filas duplicadas, lo que plantea la posibilidad de que haya datos duplicados, pues los diamantes no cuentan con un número identificador.

Al no saber a ciencia cierta si cada fila corresponde a un diamante único y teniendo en cuenta que los posibles duplicados representan tan solo un 1,48% del total se toma la decisión de conservarlos.

En el caso de que se tomara la decisión de eliminarlos, se podría realizar con la función `drop_duplicates()` de la librería _pandas_.


#4 - ANÁLISIS CORRELATIVO



##4.1 HEATMAP

Para un primer análisis de las relaciones entre las diferentes propiedades de los diamantes se va a mostrar un _heatmap_ y obtener más información.

````python
#Selecciono únicamente las columnas numéricas
numerical_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(12, 10))
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
````

En el mapa de calor podemos observar que tanto el peso como las medidas ```x```, ```y``` y ```z``` están fuertemente relacionadas con el valor del diamante.
De entre estas cuatro columnas se van a eliminar ```x```, ```y``` y ```z```, dado que no ofrecen tanta información por sí mismas como el peso del diamante.

````python
df = df.drop(['x', 'y', 'z'], axis=1)
df.head(5)
````
